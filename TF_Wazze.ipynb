```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
### Importamos los datos de Lima-intersecciones.csv y los convertimos en un DataFrame
Intersecciones = pd.read_csv('lima_intersecciones.csv', sep=';')
Intersecciones
```

```python
###Generamos la grafica a partir de la Dataframe
Waze = nx.from_pandas_edgelist(Intersecciones,source='ID_Origen_intereccion',target='ID_Final_Interseccion',edge_attr=True)
```

```python
Waze.nodes()
```

```python
Waze.edges()
```

```python
### Cantidad de vertices o nodos 
Waze.order()
```

```python
### Calles Frecuentes con mas de 1 interseccion
for x in Waze.nodes():
    if Waze.degree(x) > 2:
        print(x)
```

```python
### Ruta mas corta
djk_path=nx.dijkstra_path(Waze, source=4463 ,target=19421 ,weight=True)
djk_path
```
